## 1. Ошибки токенизации
Найдите 1 любой способ сломать токенизацию на предложения функцией sentenize из библиотеки razdel. Придумайте (или найдите на каком-то корпусе) такое предложение (или несколько предложений), которое будет некорректно разобрано sentenize, но при этом будет грамматически верным.

На natasha.github (что бы это не значило), есть памятка:
Правила в Razdel оптимизированы для аккуратно написанных текстов с правильной пунктуацией. Решение хорошо работает с новостными статьями, художественными текстами. На постах из социальных сетей, расшифровках телефонных разговоров качество ниже.

```python
from razdel import sentenize

text = """
    У нас многие люди в переписках
    Могут писать вот так
    Че уж греха таить, ставить точки в конце -- тяжело
    Я вчера купил пива
    А оно мне было как раз
"""

for item in list(sentenize(text)):
    print(item)

print("---")

text = "Сколько можно говорить :( мы пытались исправить данную ситуацию, но, к сожалению, ничего не получилось"

for item in list(sentenize(text)):
    print(item)
```

### То на то и выходит, что, чтобы сломать данную функцию, нужно всего лишь быть любителем посидеть на форумах со словами "я не на диктанте", потому что, если мы-таки поставим там заглавную, то получим:

```python
text = "Сколько можно говорить :( мы пытались исправить данную ситуацию, но, к сожалению, ничего не получилось"

for item in list(sentenize(text)):
    print(item)
```

```python
# И, кроме того, еще можно найти вот такой прикол, касающийся всяких загадочных личностей:

text = "Ну мы вчера погуляли....А потом я купил картошки и пошел домой......"

for item in list(sentenize(text)):
    print(item)
```

## 3. Лемматизация Mystem vs Pymorphy
Лемматизируйте текст с помощью mystem и pymorphy. Найдите различия в лемматизации. Что, по вашему мнению, работает лучше на этом тексте?

```python
from pymorphy3 import MorphAnalyzer
import nltk

# nltk.download("punkt")

morph = MorphAnalyzer()

mystemResult = mystem.lemmatize(textToGo)
pymorphyResult = [morph.parse(word)[0].normal_form for word in nltk.word_tokenize(textToGo)]

print(''.join(mystemResult))
print(''.join(map(lambda it: it if it == ',' or it == '.' else " " + it, pymorphyResult)))
```

оставаться -> остаться от слова "осталось"

научаться -> научиться

этот -> это

поздно -> поздний

во -> в

со -> с

В общем и целом, я вновь, как и раньше, доволен pymorphy. Кроме того, он обретает еще более прекрасные краски с токенайзером от nltk.

```python
nTest = '''
Мне мой друг вчера принес\n\n\n\n\n
Гидравлический насос
И сказал, что это ландыши
'''

list(nltk.word_tokenize(nTest))
```

И снова мы видим, что не воспринимаем мы \n как нечто разделительное, хотя в этот раз как токенайзер заюзали nltk. Тогда давайте проверим и Mystem:

```python
nTest = '''
Мне мой друг вчера принес
Гидравлический насос
И сказал, что это ландыши
'''

list(mystem.lemmatize(nTest))
```

Да, это очень занятно, только Mystem воспринимает \n как что-то значимое в контексте токенизации
